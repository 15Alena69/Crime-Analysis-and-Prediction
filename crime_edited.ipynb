# Data Preprocessing

```python
#Treating garbage values
df.loc[df['Vict Age']==0.0,'Vict Age']=np.NaN
df['DATE OCC']=df['DATE OCC'].str.replace('12:00:00 AM','')
df['Mocodes']=df['Mocodes'].str.replace(' ','')
df['Date Rptd']=df['Date Rptd'].str.replace('12:00:00 AM','')
```

```python
#Percentage of Missing Values
(df.isna().sum()/df.shape[0])*100
```

```python
df.drop(['DR_NO','Crm Cd 2','Crm Cd 3','Crm Cd 4','Cross Street','Status'],axis=1,inplace=True)
df.drop(['AREA NAME','Crm Cd Desc','Premis Desc','Weapon Desc','Status Desc'],axis=1,inplace=True)
```

```python
#Filling missing values
float_lst=['Part 1-2','Crm Cd','Vict Age','Premis Cd','Weapon Used Cd','Crm Cd 1','LAT','LON']
obj_lst=['Mocodes','Vict Sex','Vict Descent','LOCATION']
for i in float_lst:
  df[i].fillna(df[i].median(),inplace=True)
for j in obj_lst:
  df[j].fillna(df[j].mode()[0],inplace=True)
df.isna().sum()
```

```python
df.dtypes
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
end=LabelEncoder()
obj=['Date Rptd','DATE OCC','Vict Sex','Vict Descent','LOCATION']
for k in obj:
  df[k]=end.fit_transform(df[k])

df['Mocodes']=df['Mocodes'].astype(float)
df.dtypes
```

```python
#Outlier Treatment
#It is done only for continous numerical variables[not for categorical and output variables]
def wisker(col):
  q1=col.quantile(0.25)
  q3=col.quantile(0.75)
  IQR=q3-q1
  LW=q1-1.5*IQR  #lower wisker
  UW=q3+1.5*IQR  #upper wisker
  return LW,UW
```

```python
for i in ['Vict Age','Mocodes','LAT','LON']:
  LW,UW=wisker(df[i])
  df[i]=np.where(df[i]<LW,LW,df[i])
  df[i]=np.where(df[i]>UW,UW,df[i])
```

# Seperating Input Features(x) and Output Feature(y)

```python
x=df.drop(['Status_edited'],axis=1).values
x
```

```python
y=df['Status_edited'].values
y
```

```python
#Balancing
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from collections import Counter


# Define the pipeline with SMOTE for oversampling and RandomUnderSampler for undersampling
pipeline = Pipeline([
    ('over', SMOTE(sampling_strategy='auto')),  # Use 'auto' for SMOTE to automatically balance classes
    ('under', RandomUnderSampler(sampling_strategy='auto'))  # Use 'auto' for RandomUnderSampler to automatically balance classes
])

# Apply the pipeline to create balanced data
X_balanced, y_balanced = pipeline.fit_resample(x, y)

# Print the class distribution after balancing
print('Class distribution after balancing:', Counter(y_balanced))
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X_balanced,y_balanced,test_size=0.30,random_state=42)
x_train
```

# Normalization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

# Model Creation

```python
from sklearn.tree import DecisionTreeClassifier 
model=DecisionTreeClassifier(criterion='entropy',random_state=42)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

# Performance Evaluation

```python
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay,classification_report
print('Score=',accuracy_score(y_test,y_pred))
print('Report:',classification_report(y_test,y_pred))
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
lab=['Not Arrested','Arrested']
cmd=ConfusionMatrixDisplay(cm,display_labels=lab)
cmd.plot()
```
